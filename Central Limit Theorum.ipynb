### Defining the population dataset.

```python
import numpy.random as np
import seaborn as sns
import matplotlib.pyplot as plt
population_size = 1000000
population = np.rand(1000000)
```

```python
population
```

### Defining the samples from the polpulation.

```python
number_of_samples = 10000
sample_means = np.rand(number_of_samples)
```

### Select the sample size among 1,10, 20 , 30 and observe the difference between the distribution.

```python
sample_size = int(input("Please Enter the Sample Size among (1,10,20, 30) = "))
```

```python
sample_means
```

### Calculationg the mean of the samples taken from the population dataset.

```python
import numpy 
c = np.rand(number_of_samples)
for i in range(0,number_of_samples):
 c = np.randint(1,population_size,sample_size)
 sample_means[i] = numpy.mean(c)
```

```python
np.randint(1,population_size,sample_size)
```

```python
sample_means.shape
```

### Plotting the distribution of the sample mean

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means,bins=int(180/5),hist = True,kde = False)
plt.title('Histogram of Sample mea',fontsize=20)
plt.xlabel('Sample mean',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.subplot(1,2,2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means,hist = False,kde = True)
plt.title('Density of Sample mean',fontsize=20)
plt.xlabel('Sample mean',fontsize=20)
plt.ylabel('Density',fontsize=20)
plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
```

## As you increase the sample size the distribution of the sample means is getting more close to the Normal/Gaussian Distribution which is stated by the Central Limit Theorum.
